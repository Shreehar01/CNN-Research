```python
from sklearn import preprocessing
df.loc[:, df.columns!='Unnamed: 0'] = preprocessing.scale(df.loc[:, df.columns!='Unnamed: 0'], axis=0)
```

```python
df['Labels'] = df_labels['x']
```

```python
df.set_index('Labels', inplace=True)
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
Z1= linkage(df, 'complete')
fig = plt.figure(figsize=(25, 20))
fig.suptitle("Complete Linkage")
dn = dendrogram(Z1) # plotting the dendrograms
plt.show()
```

```python
Z2= linkage(df, 'single')
fig = plt.figure(figsize=(25, 10))
fig.suptitle("Single Linkage")
dn = dendrogram(Z2) # plotting the dendrograms
plt.show()
```

```python
Z3= linkage(df, 'average')
fig = plt.figure(figsize=(25, 10))
fig.suptitle("Average Linkage")
dn = dendrogram(Z3) # plotting the dendrograms
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
column_labels = fcluster(Z1, t=4, criterion='maxclust')
column_labels
```

```python
pd.crosstab(df.index, column_labels)
```

```python
from sklearn.cluster import KMeans
km1 = KMeans(n_clusters=4, n_init=150, random_state=123)
km1.fit(df)
```

```python
pd.crosstab(df.index, km1.labels_)
```
